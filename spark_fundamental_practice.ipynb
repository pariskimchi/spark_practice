```python
log_of_songs = [
        "Despacito",
        "Nice for what",
        "No tears left to cry",
        "Despacito",
        "Havana",
        "In my feelings",
        "Nice for what",
        "Despacito",
        "All the stars"
]
```

```python
play_count = 0
```

```python
def count_plays(song_title):
    global play_count
    for song in log_of_songs:
        if song == song_title:
            play_count = play_count + 1
    return play_count
```

```python
count_plays("Despacito")
```

```python
count_plays("Despacito")
```

```python
count_plays("Despacito")
```
